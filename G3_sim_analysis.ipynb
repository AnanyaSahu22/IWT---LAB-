```python
import numpy as np
import matplotlib.pyplot as plt

# Read the entire file and split it into a clean list of all numbers
with open('/ph00123.dat', 'r') as file:
    words = file.read().split()

nph_data = []
idx = 0  # Word pointer index
total_words = len(words)

event_count = 0

while idx < total_words:
    # 1. Word 2 (idx + 1) gives us the number of paddles (np)
    np_paddles = int(words[idx + 1])

    # 2. Calculate the exact word position of nph using the guide's formula
    # Subtract 1 because Python arrays start indexing at 0
    nph_word_idx = idx + (2 + 2 * np_paddles + 11) - 1

    # 3. Extract the number of photons reaching the PMT
    nph = int(words[nph_word_idx])
    nph_data.append(nph)

    # 4. Calculate total words this entire event occupies
    # (14 + 2*np + nph)
    total_event_words = 14 + 2 * np_paddles + nph

    event_count += 1
    print(f"Event {event_count}: np = {np_paddles}, nph = {nph}")

    # 5. Jump the pointer straight to the start of the next event block
    idx += total_event_words

# --- Plot the distribution ---
plt.figure(figsize=(8, 5))
plt.hist(nph_data, bins='auto', edgecolor='black', color='crimson', alpha=0.75)
plt.xlabel('Number of Photons Reaching PMT ($n_{ph}$)', fontsize=12)
plt.ylabel('Frequency (Counts)', fontsize=12)
plt.title('Distribution of Photons Reaching PMT across Simulation Events', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def parse_g3sim_energy_data(filename):
    """Parses a G3sim data file and extracts BOTH produced photons and reached photons."""
    try:
        with open(filename, 'r') as file:
            words = file.read().split()
    except FileNotFoundError:
        print(f"File {filename} not found. Skipping.")
        return None

    produced_photons = []
    reached_photons = []

    idx = 0
    total_words = len(words)

    while idx < total_words:
        if idx + 1 >= total_words:
            break

        np_paddles = int(words[idx + 1])

        # Calculate exact word positions based on the user guide layout
        produced_word_idx = idx + (2 + 2 * np_paddles + 4) - 1
        reached_word_idx = idx + (2 + 2 * np_paddles + 11) - 1

        # Safety check for unexpected simulation cut-offs
        if reached_word_idx >= total_words:
            break

        n_prod = int(words[produced_word_idx])
        n_reach = int(words[reached_word_idx])

        produced_photons.append(n_prod)
        reached_photons.append(n_reach)

        # Advance index to the start of the next event
        total_event_words = 14 + 2 * np_paddles + n_reach
        idx += total_event_words

    return np.array(produced_photons), np.array(reached_photons)

# --- Configuration ---
# Replace these strings with your actual 3 simulation filenames
energy_runs = {
    'Low Energy': '/content/ph00002.dat',
    'Medium Energy': '/content/ph01000.dat',
    'High Energy': '/content/ph00003.dat'
}

# Set up a side-by-side subplot canvas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Define distinct colors for clear comparison
colors = {'Low Energy': '#1f77b4', 'Medium Energy': '#ff7f0e', 'High Energy': '#2ca02c'}

# Process each simulation file and plot
for label, filename in energy_runs.items():
    data = parse_g3sim_energy_data(filename)

    if data is not None:
        n_prod, n_reach = data

        # Plot 1: Photons Produced in Scintillator
        ax1.hist(n_prod, bins=40, density=True, histtype='step',
                 linewidth=2.5, color=colors[label], label=label)

        # Plot 2: Photons Reached PMT
        ax2.hist(n_reach, bins=40, density=True, histtype='step',
                 linewidth=2.5, color=colors[label], label=label)

# --- Styling Left Plot (Produced) ---
ax1.set_title('Photons Produced in Scintillator', fontsize=13, fontweight='bold')
ax1.set_xlabel('Number of Photons Generated', fontsize=11)
ax1.set_ylabel('Probability Density', fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend(loc='upper right')

# --- Styling Right Plot (Reached) ---
ax2.set_title('Photons Reaching PMT ($n_{ph}$)', fontsize=13, fontweight='bold')
ax2.set_xlabel('Number of Photons Detected', fontsize=11)
ax2.set_ylabel('Probability Density', fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Reuse the parsing function from your previous script
def parse_g3sim_energy_data(filename):
    try:
        with open(filename, 'r') as file:
            words = file.read().split()
    except FileNotFoundError:
        return None

    produced, reached = [], []
    idx = 0
    total_words = len(words)

    while idx < total_words:
        if idx + 1 >= total_words: break
        np_paddles = int(words[idx + 1])
        prod_idx = idx + (2 + 2 * np_paddles + 4) - 1
        reach_idx = idx + (2 + 2 * np_paddles + 11) - 1

        if reach_idx >= total_words: break

        produced.append(int(words[prod_idx]))
        reached.append(int(words[reach_idx]))
        idx += 14 + 2 * np_paddles + int(words[reach_idx])

    return np.array(produced), np.array(reached)

energy_runs = {
    'Low Energy': '/content/ph00002.dat',
    'Medium Energy': '/content/ph01000.dat',
    'High Energy': '/content/ph00003.dat'
}

# Setup a 1-row, 2-column subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
colors = {'Low Energy': '#1f77b4', 'Medium Energy': '#ff7f0e', 'High Energy': '#2ca02c'}

for label, filename in energy_runs.items():
    data = parse_g3sim_energy_data(filename)
    if data is not None:
        n_prod, n_reach = data

        # Calculate Efficiency per event safely (avoid division by zero)
        efficiency = n_reach / np.where(n_prod == 0, 1, n_prod)

        # Left Plot: Efficiency Distribution
        ax1.hist(efficiency, bins=40, density=True, histtype='step',
                 linewidth=2.5, color=colors[label], label=f'{label} (Mean: {efficiency.mean():.3f})')

        # Right Plot: Scatter Points (subsample 200 points so it doesn't lag if data is huge)
        sample_idx = np.random.choice(len(n_prod), min(200, len(n_prod)), replace=False)
        ax2.scatter(n_prod[sample_idx], n_reach[sample_idx], color=colors[label], alpha=0.4, edgecolors='none')

        # Calculate Linear Regression Fit Line
        slope, intercept, r_value, p_value, std_err = linregress(n_prod, n_reach)
        x_fit = np.linspace(min(n_prod), max(n_prod), 100)
        y_fit = slope * x_fit + intercept
        ax2.plot(x_fit, y_fit, color=colors[label], linewidth=2, label=f'{label} Fit (Slope: {slope:.3f})')

# --- Labeling Left Graph ---
ax1.set_title('Propagation Efficiency Distribution ($N_{reach} / N_{prod}$)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Fraction of Photons Reaching PMT', fontsize=11)
ax1.set_ylabel('Probability Density', fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend(loc='upper right')

# --- Labeling Right Graph ---
ax2.set_title('Correlation Between Generated and Reached Photons', fontsize=12, fontweight='bold')
ax2.set_xlabel('Photons Produced in Scintillator', fontsize=11)
ax2.set_ylabel('Photons Reaching PMT', fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
